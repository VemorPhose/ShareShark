```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/NSE.csv')
df = df[df['instrument_type'] == 'EQUITY']
df.sort_values(by='exchange_token', inplace=True)
df.head()
```

```python
# List of NIFTY50 symbols (as of June 2025, update as needed)
nifty50_symbols = [
    'ADANIENT', 'ADANIPORTS', 'APOLLOHOSP', 'ASIANPAINT', 'AXISBANK', 'BAJAJ-AUTO', 'BAJAJFINSV', 'BAJFINANCE',
    'BEL', 'BHARTIARTL', 'CIPLA', 'COALINDIA', 'DRREDDY', 'EICHERMOT', 'ETERNAL', 'GRASIM', 'HCLTECH', 'HDFCBANK',
    'HDFCLIFE', 'HEROMOTOCO', 'HINDALCO', 'HINDUNILVR', 'ICICIBANK', 'INDUSINDBK', 'INFY', 'ITC', 'JIOFIN',
    'JSWSTEEL', 'KOTAKBANK', 'LT', 'M&M', 'MARUTI', 'NESTLEIND', 'NTPC', 'ONGC', 'POWERGRID', 'RELIANCE',
    'SBILIFE', 'SBIN', 'SHRIRAMFIN', 'SUNPHARMA', 'TATACONSUM', 'TATAMOTORS', 'TATASTEEL', 'TCS', 'TECHM',
    'TITAN', 'TRENT', 'ULTRACEMCO', 'WIPRO'
]

# Filter for NIFTY50 stocks
nifty50_df = df[df['tradingsymbol'].isin(nifty50_symbols)]

# Save to new CSV
nifty50_df.to_csv('../../data/processed/NIFTY50.csv', index=False)

print(nifty50_df.shape)
print(nifty50_df.head())
```
